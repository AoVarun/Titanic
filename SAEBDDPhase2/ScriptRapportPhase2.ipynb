### <u> **A) Création de tables(Passagers,Trajet,titanic)** </u> 

```python
%%sql

create table Passagers(
    Name varchar(200) not null,
    Sex varchar(10),
    Age float,
    SibSp int,
    Parch int,
    primary key(Name)    );

create table Trajet(
    PassengerId int not null,
    Pclass int,
    Ticket varchar(50),
    Fare float,
    Cabin varchar(20),
    primary key(PassengerId)   );

create table titanic(
    Name varchar(200) not null,
    PassengerId int not null,
    Survived int,
    Enbarked varchar(5),
    foreign key(Name) references Passagers(Name),
    foreign key(PassengerId) references Trajet(PassengerId)   );

```

# <u> **Rapport Phase 2:** </u>  <br>

##  <u> **Introduction:** </u>   

 ### &nbsp;    Nous avons commencé à créer précisément 3 tables (relations). Ces tables permettent de bien structurer et d’organiser les données. Nous avons classé les données dans les relations/tables en fonction de plusieurs  critères.


## <u> **1) Table Passagers :** </u>

&ensp;     **Dans la première Table que nous avons nommée“Passagers”, nous classons dans plusieurs attributs les informations du passager lors de la croisière du Titanic.**<br>
Nous avons 5 attributs: <br>
-Le premier attribut est **“Name”** avec le type **string**  (chaîne de caractères) qui liste tous les prénoms des passagers. <br>
-Le second attribut est **“Sex”** avec un type **str**(string/chaîne de caractères).Cette attribut correspond au sexe des passagers si il est **‘male’** ou **‘female’**. <br>
-Le troisième attribut est **“Age”**, il est de type **float**(c’est à dire un nombre à virgule).Ce sont les âges des passagers. <br>
-Le quatrième attribut est **“SibSp”** de type **integer**(valeur entière).Ici, cet attribut indique le nombre d’époux/femmes parmi les passagers. <br>
-Le cinquième attribut est **"Parch"** de type **integer** et indique le nb de parents ou enfants passagers. <br>
<br>
La **clé primaire** de cette relation “Passagers “ est l’attribut **“Name”** car elle n’a ni emplacement vide (not null) ni de tuples répétitives (c’est à dire qu’il n'y’a que des tuples uniques).



















## <u> **2) Table Trajet :** </u>

&ensp; **Dans la deuxième table que nous avons nommée “Trajet” où sont classées toutes les données portant sur le séjour des passagers en lui-même.** <br>
Nous avons aussi 5 attributs: <br>
-Le premier attribut est **“PassengerId”** de type **integer** qui quant à lui associe un nombre à chaque passager. <br>
-Le second attribut est **“Pclass”** de type **int(variant de 1 à 3)** qui quant à lui indique le niveau de classe du passager en fonction de leurs moyens. <br>
-Le troisième attribut est **“Ticket”** de type **string** qui correspond tout simplement au  numéro du passager. <br>
-Le quatrième attribut est **“Fare”** de type **float**(nombre à virgules).Cet attribut indique le prix payé par un passager pour effectuer la croisière titanic.<br>
-Le dernier attribut est **“Cabin” de type **string**(chaîne de caractère)qui  liste le(s) cabine(s) associés au passagers.<br>
<br>
L’attribut **“PassengerId”** est la **clé primaire** (elle a aucun emplacement vide et tous les tuples sont uniques). <br>









## <u> **3) Table titanic :** </u>

&ensp; **Pour finir, la dernière relation que nous avons nommé “titanic” où sont classés  les informations portant sur les embarcations et les survivants après la collision du Titanic est composée de 4 attributs:** <br>
-Le premier attribut est **“Name”** de type **string**. <br>
-Le deuxième attribut est **“PassengerId”** de type **integer**. <br>
-Le troisième attribut est **“Survived”** de type **integer(variants de 0 à 1)**.Si le **chiffre** correspond à **1**, le passager a **survécu**, si le chiffre correspond à 0, il n’a pas survécu.<br>
-Le dernier attribut est **“embarked”** de type **string** qui attribue des lettres aux embarcations.<br>
<br>
**Il n’y a pas de clés primaires** dans cette **relation “titanic”**.Cependant, il y’a bien **2 clés étrangères**(Une clé étrangère est une clé primaire dans une autre relation;ces clés primaires sont très utiles pour faire des jointures entre différentes tables/relations) dans cette relation, il y’a **“Name” qui est une clé primaire dans la table “Passagers” mais qui est donc une clé étrangère dans la table “titanic”** et il y’a aussi **“PassengerId” qui est la clé primaire dans la table “Trajet” mais qui est donc une clé étrangère dans la table “titanic”.**
